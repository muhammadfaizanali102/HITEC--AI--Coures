```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Step 1: Load the CSV file
data = pd.read_csv(r"C:\Users\dell\Downloads\canada_per_capita_income.csv")

# Inspect the data
print(data.head())
print("-" * 30)
missing_count_per_column = data.isnull().sum()
print("Missing Values Per Column:")
print(missing_count_per_column)
print("-" * 30)
print(data.describe())
any_missing = data.isnull().values.any()
print(f"Are there any missing values in the DataFrame? {any_missing}")

# Step 2: Prepare data for regression
# Assuming the CSV has columns: 'year' and 'per capita income (US$)'
X = data['year']       # Independent variable
y = data['per capita income (US$)']  # Dependent variable

print("X Data Type:", X.dtype)
print("y Data Type:", y.dtype)
X_reshaped = X.values.reshape(-1, 1)  # Converts Series to 2D array for regression

# Step 3: Create and train the model
model = LinearRegression()
model.fit(X_reshaped, y)

# Step 4: Predict per capita income for the year 2020
year_to_predict = np.array([[2020]])
predicted_income = model.predict(year_to_predict)
print(f"Predicted per capita income in 2020: ${predicted_income[0]:.2f}")

# Step 5: Optional - visualize the regression line
plt.scatter(X, y, color='blue')
plt.plot(X_reshaped, model.predict(X_reshaped), color='red', linewidth=2)
plt.scatter(2020, predicted_income, color='green', marker='o', s=100, label='2020 Prediction')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income Prediction")
plt.legend()
plt.show()
```
