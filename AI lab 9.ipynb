```python
import pandas as pd
try:
    df = pd.read_csv(r'D:\AI codes\framingham_Heart_Disease_Dataset.csv')
except FileNotFoundError:
    print("Error: framingham_Heart_Disease_Dataset.csv not found.")
    exit()
required_cols = ['sysBP', 'diaBP', 'TenYearCHD']
if all(col in df.columns for col in required_cols):
    df_clean = df.dropna(subset=required_cols)
    high_bp = (df_clean['sysBP'] >= 130) | (df_clean['diaBP'] >= 80)

    heart_disease = df_clean['TenYearCHD'] == 1
    count_both = (high_bp & heart_disease).sum()
    print(f"Number of individuals with high blood pressure and heart disease: {count_both}")
    print(f"Total individuals with valid data in these columns: {len(df_clean)}")
    total_high_bp = high_bp.sum()
    total_hd = heart_disease.sum()
    print(f"Individuals with high blood pressure only: {total_high_bp - count_both}")
    print(f"Individuals with heart disease only: {total_hd - count_both}")
else:
    print("Error: One or more required columns not found.")
    print(f"Available columns: {list(df.columns)}")
    print(f"Required: {required_cols}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
file_path = r"D:\AI codes\framingham_Heart_Disease_Dataset.csv"
df = pd.read_csv(file_path)
print(df.head())
print(df.info())
features = ['age', 'totChol', 'sysBP', 'BMI']
data = df[features].dropna()
plt.figure()
plt.scatter(data['age'], data['sysBP'])
plt.xlabel('Age')
plt.ylabel('Systolic BP')
plt.title('Age vs Systolic Blood Pressure')
plt.show()
plt.figure()
plt.scatter(data['BMI'], data['totChol'])
plt.xlabel('BMI')
plt.ylabel('Total Cholesterol')
plt.title('BMI vs Total Cholesterol')
plt.show()
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)
plt.figure()
plt.scatter(data['age'], data['sysBP'], c=clusters)
plt.xlabel('Age')
plt.ylabel('Systolic BP')
plt.title('Clustering: Age vs Systolic BP')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("D:\\AI codes\\framingham_Heart_Disease_Dataset.csv")
numeric_columns = ['age', 'BMI', 'totChol', 'sysBP', 'diaBP', 'heartRate', 'glucose']
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers, lower_bound, upper_bound
print("Outlier Detection Results (using IQR method):")
for col in numeric_columns:
    if col in df.columns:
        data = df[col].dropna()  
        outliers, lower, upper = detect_outliers_iqr(data)
        print(f"\nColumn: {col}")
        print(f"  - Number of outliers: {len(outliers)}")
        print(f"  - Outlier values: {outliers.tolist()[:10]}...")  
        print(f"  - Bounds: Lower = {lower:.2f}, Upper = {upper:.2f}")
    else:
        print(f"\nColumn: {col} not found.")

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()
for i, col in enumerate(numeric_columns):
    if col in df.columns and i < len(axes):
        df.boxplot(column=col, ax=axes[i])
        axes[i].set_title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```
